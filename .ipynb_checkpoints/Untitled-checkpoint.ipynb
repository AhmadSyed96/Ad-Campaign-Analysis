# Marketing-Ad-Campaign Analysis

the marketing team has alerted us to a problem with our campaign, lets investigate the data

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
marketing = pd.read_csv('marketing.csv')
marketing['date_served'] = pd.to_datetime(marketing.date_served, infer_datetime_format=True)
marketing['date_subscribed'] = pd.to_datetime(marketing.date_subscribed, infer_datetime_format=True)
marketing['date_canceled'] = pd.to_datetime(marketing.date_canceled, infer_datetime_format=True)
marketing
```

lets set up functions to be used throughout

```python
def conversion_rate(df, cols):
    subscribers = df[df.converted == True].groupby(cols)['user_id'].nunique()
    total = df.groupby(cols)['user_id'].nunique()
    conv_rt = subscribers / total
    return conv_rt
```

```python
def plotting_conv(df):
    for col in df:
        plt.plot(df.index, df[col])
        plt.title('Daily '+str(col)+' conversion rate \n', size=16)
        plt.ylabel('Conversion Rate (%)', size=14)
        plt.xlabel('Date',size=14)
        plt.xticks(rotation=45)
        plt.show()
        plt.clf()
```

lets see which marketing channel this issue is affecting

```python
daily_conversion_rate_by_channel = conversion_rate(marketing, ['date_served', 'marketing_channel'])
daily_conversion_rate_by_channel_df = pd.DataFrame(daily_conversion_rate_by_channel.unstack(level=1))
plotting_conv(daily_conversion_rate_by_channel_df)
```

it seems house ads has experienced as sharp decline in conversion rate, lets investigate what could be causing this

```python
# Add day of week column to marketing
marketing['DoW_served'] = marketing.date_served.dt.dayofweek

# Calculate conversion rate by day of week
DoW_conversion = conversion_rate(marketing, ['DoW_served', 'marketing_channel'])


# Unstack channels
DoW_df = pd.DataFrame(DoW_conversion.unstack(level=1))
print(DoW_df)

# Plot conversion rate by day of week
DoW_df.plot()
plt.title('Conversion rate by day of week\n')
plt.ylim(0)
plt.show()
```

day of week seems to have no affect house ads, lets check the language it was presented in

```python
house_ads = marketing[marketing['marketing_channel']=='House Ads']
house_ads_conversions_by_language = conversion_rate(house_ads, ['date_served','language_displayed'])
house_ads_conversions_by_language_df = pd.DataFrame(house_ads_conversions_by_language.unstack(level=1))
plotting_conv(house_ads_conversions_by_language_df)
# house_ads_conversions_by_language_df
```

it seems english had a drop after the 11th, corresponding to the other languages. lets check the ratio of correct language served per day

```python
house_ads['is_correct_lang'] = np.where(house_ads.language_displayed==house_ads.language_preferred,'yes','no')
language_check = house_ads.groupby(['date_served','is_correct_lang'])['user_id'].count()
language_check_df = pd.DataFrame(language_check.unstack(level=1)).fillna(0)
language_check_df
```

```python
language_check_df['pct'] = language_check_df['yes'] / language_check_df.sum(axis=1)
plt.plot(language_check_df.index, language_check_df['pct'])
plt.xticks(rotation=45)
plt.show()
```

now that we know when the error occured and that english was not effected, we see how many conversions were lost due to this error by comparing how other languages fared before the error and applying that for the time period the error occured 

```python
# calculate the ratios compared to english for the period before the error
house_ads_no_bug = house_ads[house_ads['date_served']<'2018-01-11']
lang_conv = conversion_rate(house_ads_no_bug, ['language_displayed'])
spanish_index = lang_conv['Spanish'] / lang_conv['English']
arabic_index = lang_conv['Arabic'] / lang_conv['English']
german_index = lang_conv['German'] / lang_conv['English']

# calculate how many subscribers there should have been for each language based on the ratio for the time during the error
lang_conv2 = house_ads.groupby(['date_served','language_preferred']).agg({'user_id':'nunique','converted':'sum'})
lang_conv_df = pd.DataFrame(lang_conv2.unstack(level=1))
lang_conv_df['actual_english_rate'] = lang_conv_df['2018-01-11':'2018-01-31'][('converted','English')]
lang_conv_df['expected_spanish_rate'] = lang_conv_df.actual_english_rate * spanish_index
lang_conv_df['expected_spanish_conversion'] = lang_conv_df['expected_spanish_rate'] / 100 * lang_conv_df[('user_id','Spanish')]
lang_conv_df['expected_arabic_rate'] = lang_conv_df.actual_english_rate * arabic_index
lang_conv_df['expected_arabic_conversion'] =  lang_conv_df['expected_arabic_rate'] / 100 *lang_conv_df[('user_id','Arabic')]
lang_conv_df['expected_german_rate'] = lang_conv_df.actual_english_rate * german_index
lang_conv_df['expected_german_conversion'] =  lang_conv_df['expected_german_rate'] / 100 *lang_conv_df[('user_id','German')]

# add up all of the lost subscribers
bug_period = lang_conv_df.loc['2018-01-11':'2018-01-31']
expected_subs = bug_period['expected_spanish_conversion'].sum() + bug_period['expected_arabic_conversion'].sum() + bug_period['expected_german_conversion'].sum()
actual_subs = bug_period[('converted','Spanish')].sum() + bug_period[('converted','Arabic')].sum() + bug_period[('converted','German')].sum()
lost_subs = expected_subs - actual_subs
lost_subs
```
